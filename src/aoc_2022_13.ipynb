When comparing two values, the first value is called left and the second value
is called right. Then:

If both values are integers, the lower integer should come first.   
* If the left integer is lower than the right integer, the inputs are in   
    the right order.    
* If the left integer is higher than the right integer, the inputs are not 
    in the right order.    
* Otherwise, the inputs are the same integer; continue checking the next 
    part of the input.   
 
If both values are lists, compare the first value of each list, then the second
value, and so on. 
* If the left list runs out of items first, the inputs are in the right
order. 
* If the right list runs out of items first, the inputs are not in the
right order. 
* If the lists are the same length and no comparison makes a decision 
about the order, continue checking the next part of the input.

If exactly one value is an integer, convert the integer to a list which contains
that integer as its only value, then retry the comparison. For example, if
comparing `[0,0,0]` and `2`, convert the right value to `[2]` (a list 
containing `2`); the result is then found by instead comparing `[0,0,0]` and
`[2]`.

To determine whether pairs of packets are in the right order, you can use the following algorithm:

If both values are integers, compare the integers. If the left integer is lower than the right integer, the inputs are in the right order. If the left integer is higher than the right integer, the inputs are not in the right order. Otherwise, the inputs are the same integer; continue checking the next part of the input.
If both values are lists, compare the first value of each list, then the second value, and so on. If the left list runs out of items first, the inputs are in the right order. If the right list runs out of items first, the inputs are not in the right order. If the lists are the same length and no comparison makes a decision about the order, continue checking the next part of the input.
If exactly one value is an integer, convert the integer to a list which contains that integer as its only value, then retry the comparison. For example, if comparing [0,0,0] and 2, convert the right value to [2] (a list containing 2); the result is then found by instead comparing [0,0,0] and [2].
To determine the number of pairs of packets that are in the right order, you can use the above algorithm to compare each pair of packets and count the number of times the inputs are in the right order.
```python
import ast

# Function that compares two values and returns True if the left value is smaller
# than the right value, or False otherwise
def compare_values(left, right):
    print("==>",left, type(left))
    print("==>",right, type(right))
    print("Comparing Values")
    
    # If both values are integers, compare them directly
    if isinstance(left, int) and isinstance(right, int):
        #print("1 compare integers", left < right)
        return left < right
    # If both values are lists, compare the first value of each list, then the
    # second value, and so on
    elif isinstance(left, type(list)) and isinstance(right, type(list)):
        #print("Both elements are lists")
        for l, r in zip(left, right):
            # If one of the values is smaller, return True or False accordingly
            if compare_values(l, r):
                #print("3: compare sub elements")
                return True
            elif compare_values(r, l):
                #print("4 compare inverted sub elements")
                return False
            
            # If all values are the same, return True if the left list is shorter than
            # the right list, or False otherwise
            
        #print("5:All Values are thes same")
        return len(left) < len(right)
        # If exactly one value is an integer, convert the integer to a list containing
        # the integer and retry the comparison
    elif isinstance(left, int):
        #print("6 Convert right  integer to list and compare")
        return compare_values([left], right)
    elif isinstance(right, int):
        #print("7 convert left integer to list and compare")
        return compare_values(left, [right])
    else:
        print("8 I have no idea what happened here")

# Function that compares two packets and returns True if the left packet is
# in the right order relative to the right packet, or False otherwise
def compare_packets(pair_of_packets):
    
    left = pair_of_packets[0]
    right = pair_of_packets[1]
    
    print("Converting")
    # Convert the packets to lists
    left = ast.literal_eval(left)
    right = ast.literal_eval(right)
    # Use the compare_values() function to compare the two lists
    return compare_values(left, right)

pair = []
i = -1
s = 0
for d in data:
    print(d)
    pair.append(d) 
    i += 1
    if (d == "" or i == 1): 
        # evaluate pair.
        s += compare_packets(pair) + 0
        print(i, pair, s)
        pair = []
        i = -1
    
print(i, s)
```
